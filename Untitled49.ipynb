```python
import os
from os import listdir
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
import pickle
import pathlib
import PIL
import PIL.Image
import cv2

dir = 'C:/Users/farsharb/Desktop/ML/ML_NEW/CatsDogs'


img_width = 50
img_height = 50
img_size = (img_width, img_height)
channels = 1

pets = []
categories = ['Cats','Dogs']

# funciotn in order to load the data, transform the into array, assign them to a class and store them in the pets list
def create_data():
    for category in categories:
        path = os.path.join(dir, category)
# assign 0 to cat and 1 to dog, according to the index of the categories
        pet_class = categories.index(category)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
                new_array = cv2.resize(img_array, img_size)
                pets.append([new_array, pet_class])
            except:
                pass

create_data()

print(len(pets))

random.shuffle(pets)

for sample in pets[:9]:
    print(sample[1])

# split the data arrays and the labels for the classification task

X = []
y = []

for image, label in pets:
    X.append(image)
    y.append(label)

X = np.array(X).reshape(-1, img_width, img_height, channels)

pickle_out = open('C:/Users/farsharb/Desktop/ML/ML_NEW/CatsDogs/Pickles/X.pickle','wb')
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open('C:/Users/farsharb/Desktop/ML/ML_NEW/CatsDogs/Pickles/y.pickle','wb')
pickle.dump(y, pickle_out)
pickle_out.close()

pickle_in = open('C:/Users/farsharb/Desktop/ML/ML_NEW/CatsDogs/Pickles/X.pickle','rb')
X = pickle.load(pickle_in)

pickle_in = open('C:/Users/farsharb/Desktop/ML/ML_NEW/CatsDogs/Pickles/y.pickle','rb')
y = pickle.load(pickle_in)
```

# Model 1

```python
import tensorflow as tf
from tensorflow.keras import layers
from sklearn.model_selection import KFold
from sklearn.metrics import zero_one_loss

# define the number of splits for cross-validation
k = 5

# create a KFold object
kf = KFold(n_splits=k, shuffle=True, random_state=42)

# initialize a list to store the validation accuracy for each fold
val_accs = []

# loop over the folds
for i, (train_index, val_index) in enumerate(kf.split(X)):
    print(f"Fold {i+1}/{k}")
    
    # split the data into training and validation sets for this fold
    X_train, y_train = X[train_index], y[train_index]
    X_val, y_val = X[val_index], y[val_index]
    
    # define the model architecture
    model = tf.keras.Sequential([
        layers.Conv2D(64, (3, 3), padding = "same", use_bias=False, input_shape = (50, 50, 1), activation='relu'),
        layers.Conv2D(64, (3, 3), padding = "same", use_bias=False, activation='relu'),
        layers.MaxPooling2D(pool_size = (2, 2)),
        layers.Flatten(),
        layers.Dense(128, use_bias=False, activation='relu'),
        layers.Dense(1, activation = 'sigmoid')
    ])
    
    # compile the model
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    
    # train the model
    model.fit(X_train, y_train, epochs=2, batch_size=128, validation_data=(X_val, y_val))
    
    # evaluate the model on the validation set
    val_loss, val_acc = model.evaluate(X_val, y_val)
    val_accs.append(val_acc)

# calculate the mean and standard deviation of the validation accuracy across folds
mean_val_acc = np.mean(val_accs)
std_val_acc = np.std(val_accs)
print(f"Mean validation accuracy: {mean_val_acc:.4f} +/- {std_val_acc:.4f}")

# evaluate the model on the test set using zero-one loss
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
test_predictions = np.round(model.predict(X_test))
test_error = zero_one_loss(y_test, test_predictions)
print(f"Test zero-one loss: {test_error:.4f}")
```

```python
K=5
kf = KFold(n_splits=k, shuffle=True, random_state=42)
# initialize empty lists to store the training and validation losses and accuracies for each fold
train_losses_dropout = []
train_accs_dropout = []
val_losses_dropout = []
val_accs_dropout = []


for i, (train_index, val_index) in enumerate(kf.split(X)):
    print(f"Fold {i+1}/{k}")
    
    # split the data into training and validation sets for this fold
    X_train, y_train = X[train_index], y[train_index]
    X_val, y_val = X[val_index], y[val_index]
    
    # define the model architecture
    model_dropout= tf.keras.Sequential([
        layers.Conv2D(64, (3, 3), padding = "same", use_bias=False, input_shape = (50, 50, 1), activation='relu'),
        layers.Conv2D(64, (3, 3), padding = "same", use_bias=False, activation='relu'),
        layers.MaxPooling2D(pool_size = (2, 2)),
        layers.Dropout(0.25),
        layers.Flatten(),
        layers.Dense(128, use_bias=False, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(1, activation = 'sigmoid')
        ])
    
    # compile the model
    model_dropout.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    
    # train the model and store the history
    history = model_dropout.fit(X_train, y_train, epochs=3, batch_size=128, validation_data=(X_val, y_val))
    
    # append the training and validation losses and accuracies to the lists for this fold
    train_losses_dropout.append(history.history['loss'])
    train_accs_dropout.append(history.history['accuracy'])
    val_losses_dropout.append(history.history['val_loss'])
    val_accs_dropout.append(history.history['val_accuracy'])
```
